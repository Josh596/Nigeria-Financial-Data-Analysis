```python
presidential_tenures_from_2000 = {
    '2000': 'Olusegun Obasanjo (1st term)',
    '2001': 'Olusegun Obasanjo (1st term)',
    '2002': 'Olusegun Obasanjo (1st term)',
    '2003': 'Olusegun Obasanjo (2nd term)',
    '2004': 'Olusegun Obasanjo (2nd term)',
    '2005': 'Olusegun Obasanjo (2nd term)',
    '2006': 'Olusegun Obasanjo (2nd term)',
    '2007': "Umaru Musa Yar'Adua (1st term)",
    '2008': "Umaru Musa Yar'Adua (1st term)",
    '2009': "Umaru Musa Yar'Adua (1st term)",
    '2010': "Goodluck Jonathan (1st term)",
    '2011': "Goodluck Jonathan (1st term)",
    '2012': "Goodluck Jonathan (1st term)",
    '2013': "Goodluck Jonathan (1st term)",
    '2014': "Goodluck Jonathan (1st term)",
    '2015': "Muhammadu Buhari (1st term)",
    '2016':  "Muhammadu Buhari (1st term)",
    '2017':  "Muhammadu Buhari (1st term)",
    '2018':  "Muhammadu Buhari (1st term)",
    '2019':  "Muhammadu Buhari (2nd term)",
    '2020':  "Muhammadu Buhari (2nd term)",
    '2021':  "Muhammadu Buhari (2nd term)",
    '2022': "Muhammadu Buhari (2nd term)"
}

def add_president_to_series(ser:pd.Series) -> pd.DataFrame:
    df1 = ser.to_frame()
    df1.reset_index(inplace=True)
    df1.columns = ['Index', df1.columns[1]]
    df1['President'] = df1['Index'].apply(lambda x: presidential_tenures_from_2000.get(x))
    df1.set_index('Index', drop=True, inplace=True)
    return df1

# def add_president_to_dataframe(df):
#     pass
```

```python
gdp_df = add_president_to_series(gdp_data_from_2000)
```
